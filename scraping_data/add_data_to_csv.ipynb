```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
import json
import os

# Function to initialize the WebDriver
def init_driver(headless=False):
    options = webdriver.ChromeOptions()
    if headless:
        options.add_argument("--headless")  # Run in headless mode if specified
    options.add_argument("--disable-gpu")
    options.add_argument("--no-sandbox")
    return webdriver.Chrome(options=options)

# Function to retry page load on network issues
def safe_get(driver, url, retries=3, delay=5):
    for attempt in range(retries):
        try:
            driver.get(url)
            time.sleep(delay)  # Allow page to load
            return True
        except Exception as e:
            print(f"Network issue on attempt {attempt + 1}/{retries}: {e}")
            if attempt == retries - 1:
                raise Exception("Failed to load page after multiple attempts.")
    return False

# Function to scroll and detect new content
def scroll_and_collect(driver, max_attempts=10, scroll_pause_time=2):
    unique_cars = set()
    scroll_attempts = 0

    while scroll_attempts < max_attempts:
        # Parse the page source with BeautifulSoup
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        cars = soup.find_all('div', class_='list_items_post')  # Adjust the class name if needed

        # Add unique car entries to the set
        for car in cars:
            try:
                name_tag = car.find('a', class_='position_relative d_block')  # Adjust class name
                location = car.find('p', class_='truncate gr-p-date-loc').text.split('•')[-1].strip()
                if name_tag and name_tag.has_attr('href'):
                    car_name = name_tag.text.strip()
                    car_link = name_tag['href']
                    car_entry = {"name": car_name, "link": car_link, "location": location}
                    unique_cars.add(json.dumps(car_entry))  # Store as JSON string for uniqueness
            except Exception as e:
                print(f"Error processing car: {e}")
                continue

        print(f"Loaded {len(unique_cars)} unique cars so far.")

        # Scroll down and wait
        initial_car_count = len(cars)
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(scroll_pause_time)

        # Check if new content loaded
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        new_cars = soup.find_all('div', class_='list_items_post')  # Adjust the class name if needed

        if len(new_cars) > initial_car_count:
            scroll_attempts = 0  # Reset attempts if new content is loaded
        else:
            scroll_attempts += 1
            print(f"No new content. Attempt {scroll_attempts}/{max_attempts}")

    print("Scrolling completed.")
    return [json.loads(car) for car in unique_cars]  # Convert JSON strings back to dicts

# Main scraping function
def scrape_car_details(url, file_save_path, headless=False):
    driver = init_driver(headless=headless)  # Set headless mode

    try:
        # Load the URL with retries
        if not safe_get(driver, url):
            print("Failed to load the page. Exiting...")
            return

        # Scroll and collect data
        car_data = scroll_and_collect(driver)

        # Save data to JSON file
        os.makedirs(os.path.dirname(file_save_path), exist_ok=True)  # Ensure directory exists
        if(len(car_data) > 0):
            with open(file_save_path, 'w', encoding='utf-8') as f:
                json.dump(car_data, f, ensure_ascii=False, indent=4)

        print(f"Saved {len(car_data)} unique cars to {file_save_path}")
    finally:
        driver.quit()

# Main function to iterate over car models
def main():
    models = '''
        MG JETOUR Lexus Mazda Hyundai Kia Honda Nissan Ford Land Rover Mercedes-Benz Chevrolet Porsche Mini Mitsubishi Daihatsu  
        Volkswagen Subaru Jaguar Suzuki Acura AEOLUS AION AIQAR AITO Aston Martin Audi AVATR BAIC BAW
        Bentley BESTUNE BMW Buick BYD Cadillac CHANGAN Chrysler Daewoo DENZA DFSK Dodge Ferrari FIAT
        FORTHING Foton GAC GEELY Great Wall GTV Haval Hawtai Motor HONGQI HUMMER iCar Infiniti Isuzu
        JAC Jeep JMC Lamborghini LDV Leapmotor Li Maserati MAXUS McLaren ORA Peugeot Ram Rolls-Royce
        SERES Smart SOUEAST SsangYong Tesla VENUCIA Volvo WULING Xiaomi XPENG ZEEKR Sedan Hatchback
        Pickup SUV Convertible MPV Sports Station Wagon Other
    '''
    models = [model.strip().lower() for model in models.split() if model.strip()]

    for model in models:
        url = f'https://www.khmer24.com/c-cars-for-sale?keyword=&ad_field={model}'
        file_name = f'./json1/{model}.json'
        print(f"Scraping {model} from {url}...")
        scrape_car_details(url, file_name, headless=False)  # Set headless=False to see the browser

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import os

# Directory containing JSON files
json_dir = './json1/'

# Initialize an empty list to store DataFrames
dataframes = []

# Read each JSON file and append to the list
for file in os.listdir(json_dir):
    if file.endswith('.json'):
        file_path = os.path.join(json_dir, file)
        df = pd.read_json(file_path)  # Read JSON file into a DataFrame
        dataframes.append(df)

# Concatenate all DataFrames
combined_df = pd.concat(dataframes, ignore_index=True)

# Drop duplicate rows
cleaned_df = combined_df.drop_duplicates()

# Save the cleaned DataFrame back to JSON if needed
cleaned_df.to_json('./cleaned_data.json', orient='records', lines=True)

# Print the first few rows of the cleaned DataFrame
print(cleaned_df.head())
```
